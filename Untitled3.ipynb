```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("vendas_modificado.csv")

print(df['valor'].head(10))

df['valor'] = df['valor'].astype(str).apply(lambda x: re.sub(r'[^\d,.-]', '', x))

df['valor'] = df['valor'].str.replace(',', '.')

df['valor'] = df['valor'].apply(lambda x: x if re.match(r'^\d+(\.\d{1,2})?$', x) else np.nan)

df['valor'] = pd.to_numeric(df['valor'], errors='coerce')

print(df['valor'].head(10))
print("\nValores não convertidos (NaN):", df['valor'].isnull().sum())
```
